## âœ… Step 1: Import Libraries
We will use:
- `pandas` for data manipulation
- `seaborn` for visualization and loading the dataset
- `matplotlib.pyplot` for additional plotting control

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## âœ… Step 2: Load the Iris Dataset
We'll load the Iris dataset directly using seaborn's built-in dataset loader.


```python
df = sns.load_dataset("iris")
```

## âœ… Step 3: Basic Data Inspection
Let's explore:
- The shape (rows and columns)
- The column names
- The first 5 records


```python
print("Shape of the dataset:", df.shape)
print("Column names:", df.columns.tolist())
print("\nFirst 5 rows:")
print(df.head())
```

## âœ… Step 4: Summary Statistics
We'll use:
- `.info()` to understand data types and null values
- `.describe()` for numerical statistics

```python
print("Dataset Info:")
print(df.info())

print("\nDescriptive Statistics:")
print(df.describe())
```

## âœ… Step 5: Scatter Plot â€” Petal Length vs Petal Width
This plot helps us understand relationships between two features, colored by species.


```python
sns.scatterplot(data=df, x="petal_length", y="petal_width", hue="species")
plt.title("Petal Length vs Petal Width by Species")
plt.show()
```

## âœ… Step 6: Pair Plot â€” All Features
This gives a matrix of scatter plots showing all pairwise relationships.


```python
sns.pairplot(df, hue="species")
plt.suptitle("Pairplot of Iris Dataset", y=1.02)
plt.show()
```

## âœ… Step 7: Histograms
Visualize the distribution of values for each numeric feature.


```python
df.hist(figsize=(10, 8), bins=20)
plt.suptitle("Histograms of All Features")
plt.tight_layout()
plt.show()
```

## âœ… Step 8: Box Plots
Box plots help detect outliers and compare feature distributions across species.


```python
# Individual box plot example
sns.boxplot(data=df, x="species", y="sepal_width")
plt.title("Boxplot of Sepal Width by Species")
plt.show()
```

```python
# Loop through all features (excluding 'species') and plot
for col in df.columns[:-1]:
    sns.boxplot(data=df, x="species", y=col)
    plt.title(f"Boxplot of {col.title()} by Species")
    plt.show()
```

We:
- Loaded the Iris dataset using seaborn
- Explored basic structure and summary statistics using pandas
- Visualized:
  - Feature relationships (scatter, pair plots)
  - Value distributions (histograms)
  - Outliers (box plots)

This builds foundational skills in EDA (Exploratory Data Analysis) using pandas, matplotlib, and seaborn.
